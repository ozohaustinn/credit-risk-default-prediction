#Splitting the dataset

```python
predictor = df.loc[:,df.columns != "default.payment.next.month"]
target = df.loc[:,df.columns == "default.payment.next.month"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(predictor, target, random_state = 1, test_size = 0.30)
print("X_train", x_train.shape)
print("Y_train", y_train.shape)
print("X_test", x_test.shape)
print("Y_test", y_test.shape)
```

#Model training

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(x_train,y_train)
y_pred = logr.predict(x_test)
```

```python
y_pred
```

```python
#Evaluate Model
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_pred,y_test)
cnf_matrix
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure(figsize = (6,4))
plt.plot(fpr,tpr)
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

#Tuning the model

```python
target.value_counts(normalize = True)
```

```python
#using class weights
model = LogisticRegression(class_weight = "balanced", max_iter = 1000)
model.fit(x_train,y_train)
y_proba = model.predict_proba(x_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize = (6,4))
plt.plot(fpr,tpr)
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

auc_score = roc_auc_score(y_test, y_proba)
print("AUC:",round(auc_score,3))

ks_stat = max(tpr - fpr)
print("KS Statistic:", round(ks_stat,3))
```

```python
from sklearn.metrics import recall_score

thresholds = np.arange(0.05,0.5,0.05)

for t in thresholds:
    y_prediction = (y_proba >= t).astype(int)
    recall= recall_score(y_test,y_prediction)
    print(f"Threshold {t:.2f} | Recall (Default): {recall:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
threshold = 0.4
y_final = (y_proba >= threshold).astype(int)
print(confusion_matrix(y_test,y_final))
print(classification_report(y_test,y_final))
```

```python
df_test = x_test.copy()
df_test["PD"] = y_proba

df_test["risk_band"] = pd.qcut(
    df_test["PD"],
    q=5,
    labels=["Very Low", "Low", "Medium", "High", "Very High"])
df_test.groupby("risk_band")["PD"].mean()
```
